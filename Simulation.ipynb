```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,15))
k=1000
for i in [0,1]:
    for j in [0,1]:
        arr=np.random.rand(k,2)
        inside=arr[np.sqrt(arr[:,0]**2 + arr[:,1]**2)<1]
        axes[i][j].scatter(arr[:,0],arr[:,1],s=0.5,c="red")
        axes[i][j].scatter(inside[:,0],inside[:,1],s=0.5,c="blue")
        axes[i][j].set_title(4*(len(inside)/len(arr)))
        k=k*10   
```
